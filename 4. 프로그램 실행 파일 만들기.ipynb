## URL 주소 저장하기

````python
import os, re
from bs4 import BeautifulSoup as bs

import urllib.request as ur
````

````python
soup = bs(ur.urlopen('https://news.daum.net/').read(), 'html.parser')
soup
````

````python
os.chdir(r'C:\Users\HeeKwan\Desktop\Study_Of_Crawling')
f = open('links.txt', 'w')
for s in soup.find_all('div', {"class" : "item_issue"}):
    f.write(s.find_all('a')[0].get('href') + '\n')
    
f.close()
````

## 기사 본문을 파일로 저장하기

````python
article1 = 'https://news.v.daum.net/v/20210120121906457'

soup = bs(ur.urlopen(article1).read(), 'html.parser')
f = open('article1.txt', 'w')

for p in soup.find_all('p'):
    f.write(p.text + '\n')
    
f.close()
````

## 기사 제목, 본문, 하이퍼링크를 파일로 저장하기

````python
url = 'https://news.daum.net/'
soup = bs(ur.urlopen(url).read(), 'html.parser')
f = open('article_total.txt', 'w')

for d in soup.find_all('div', {"class" : "item_issue"}):
    try:
        f.write(d.text + '\n')
        f.write(d.find_all('a')[0].get('href') + '\n')
        soup2 = bs(ur.urlopen(d.find_all('a')[0].get('href')).read(), 'html.parser')
        for p in soup2.find_all('p'):
            f.write(p.text)
    except:
        pass

f.close()
````
