## 원하는 영역에서 하이퍼링크 모두 추출하기

```python
news = 'https://news.daum.net/'
soup = bs(ur.urlopen(news).read(), 'html.parser')
```

### 1. find_all로 `<a>` 태그 추출하기

```python
for q in soup.find_all('div', {"class" : "item_issue"}):
    print(q.find_all('a'))
```

### 2. get으로 href 속성값 구하기

```python
# get은 특정 태그에서 속성값을 가져오는 것을 주의하자! 
# cf) find_all은 결과값으로 여러 태그가 들어 있는 리스트를 출력한다.
for q in soup.find_all('div', {"class" : "item_issue"}):
    print(q.find_all('a').get('href'))
```

```python
for q in soup.find_all('div', {"class" : "item_issue"}):
    print(q.find_all('a')[0].get('href'))
```

하이퍼링크를 추출하는 과정

1. soup 객체에서 class 속성값이 'item_issue'인 `<div>` 태그를 find_all로 가져온다.
2. `<div>` 태그 안에서 `<a>` 태그를 find_all로 가져온다.
3. `<a>` 태그의 href 속성값을 get으로 출력한다.

* find_all 은 태그를 추출할 때 사용
* get은 태그 안에 있는 속성의 값을 추출할 때 사용

## 기사 제목과 내용 한꺼번에 추출하기

### 1. 웹 문서를 BeautifulSoup 객체에 저장하기

```python
article1 = 'https://news.v.daum.net/v/20210120120156902'
soup1 = bs(ur.urlopen(article1).read(), 'html.parser')
soup1
```

### 2. 기사 내용 가져오기

```python
for p in soup1.find_all('p'):
    print(p.text)
```

### 3. 기사 제목 가져오기

```python
headline = soup.find_all('div', {"class" : "item_issue"})
print(headline[0].text)

# headline = soup1.find_all('h3', {"class" : "tit_view"})
# print(headline[0].text)
```

## 하이퍼링크된 모든 기사의 제목과 본문 추출하기

```python
for h in headline:
    print(h.text, '\n')
```

```python
for h in headline:
    print(h.text, '\n')
    soup2 = bs(ur.urlopen(h.find_all('a')[0].get('href')).read(), 'html.parser') # 해당 기사가 올라와 있는 웹 사이트를 열어 soup2 객체에 저장
    for p in soup2.find_all('p'):
        print(p.text)
```
